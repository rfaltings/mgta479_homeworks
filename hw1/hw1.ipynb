```python
df_wtp = pl.read_csv("hw1_data.csv")

cost_per_machine = 6000
fixed_cost = 1e6

# Plotting function - Histogram
def plot_histogram(data, column, bins=10, title="", xlabel="", ylabel="Frequency"):
    # If bins is an array (bin edges), use breaks parameter; otherwise use bins
    if isinstance(bins, np.ndarray):
        p = (
            ggplot(data, aes(x=column))
            + geom_histogram(breaks=bins, color='white')
            + labs(title=title, x=xlabel, y=ylabel)
            + theme_minimal()
            + theme(figure_size=(10, 6))
        )
    else:
        p = (
            ggplot(data, aes(x=column))
            + geom_histogram(bins=bins, color='white')
            + labs(title=title, x=xlabel, y=ylabel)
            + theme_minimal()
            + theme(figure_size=(10, 6))
        )
    return p

# Plotting function - Line Chart
def plot_line_chart(data, x_column, y_column, title="", xlabel="", ylabel=""):
    p = (
        ggplot(data, aes(x=x_column, y=y_column))
        + geom_line()
        + geom_point()
        + labs(title=title, x=xlabel, y=ylabel)
        + theme_minimal()
        + theme(figure_size=(10, 6))
    )
    return p
```
