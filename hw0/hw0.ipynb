**Question 3:** Plot the cumulative distribution function (CDF) of the `example` column from the DataFrame.

```python
# You can use the following function to plot the CDF. Read through it carefully to understand how it works.
def plot_cdf(dataframe, column_name):
    data = dataframe[column_name].to_numpy()
    sorted_data = np.sort(data)
    yvals = np.arange(1, len(sorted_data)+1) / float(len(sorted_data))
    
    plot_df = pl.DataFrame({'value': sorted_data, 'cumulative_probability': yvals})
    
    p = (
        ggplot(plot_df, aes(x='value', y='cumulative_probability'))
        + geom_line()
        + labs(x='Value', y='Cumulative Probability', title='Cumulative Distribution Function (CDF)')
        + theme_minimal()
    )
    return p
```
